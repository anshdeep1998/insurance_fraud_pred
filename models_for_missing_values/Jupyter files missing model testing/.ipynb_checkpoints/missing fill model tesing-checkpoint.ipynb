```python
import pickle   
import os
import pandas as pd
```

```python
class fill_missing_values_model:
    def __init__(self ,model_missing_folder_path, data_path, column_name):
        self.model_missing_folder_path = model_missing_folder_path
        self.column_name = column_name
        self.data_path = data_path
        
    def read_data(self):
        return pd.read_csv(self.data_path)
        
    def check_same_elements_list(self):
        data = self.read_data()
        series = data[self.column_name].value_counts()
        l=[]
        for i in series:
            l.append(i)
        result = all(element == l[0] for element in l)
        return result
    
#     def load_missing_model(self):
#         os.chdir(self.model_missing_folder_path)
#         model = pickle.load(open(self.column_name + ".pickle", "rb"))
#         return model
```
