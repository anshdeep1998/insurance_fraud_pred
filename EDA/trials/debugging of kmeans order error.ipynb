# So all the problem was there in the scaling of numeric features as in training file, it was done after training and saving Kmeans clustering model and in prediction file, it was done before using the already daved Kmeans clustering model which was trained in training file and since scaling changes order of columns, when we loaded saved trained Kmeans model, it could not identify order of columns.

# So the solution was to shift the scaling process before the clustering in training file such that it is trained to cluster already scaled data so that it can cluster easily the prediction data.
